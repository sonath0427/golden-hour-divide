```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# 1. Load your Phase1_H1 CSV
df = pd.read_csv('Phase1_H1.csv')

# 2. Perform Pearson Correlation Analysis
# This tests the relationship between Hospital Density and the Spatial Gap (Gd)
r_val, p_val = pearsonr(df['Hospital_Density'], df['Gd'])

print(f"--- Statistical Results for H1 ---")
print(f"Correlation Coefficient (r): {r_val:.4f}")
print(f"P-Value: {p_val:.4f}")

# 3. Create the Validation Plot
plt.figure(figsize=(10, 6))
sns.regplot(data=df, x='Hospital_Density', y='Gd',
            scatter_kws={'s':100, 'alpha':0.6},
            line_kws={"color": "red"})

# Annotate the plot with the statistical results
plt.title('H1 Validation: Impact of Hospital Density on Spatial Gap (Gd)', fontsize=14)
plt.xlabel('Hospital Density (Facilities per sq km)', fontsize=12)
plt.ylabel('Spatial Gap (Gd %)', fontsize=12)
plt.text(df['Hospital_Density'].min(), df['Gd'].max(),
         f'r = {r_val:.2f}\np = {p_val:.4f}',
         fontsize=12, bbox=dict(facecolor='white', alpha=0.5))

plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig('H1_Validation_Plot.png', dpi=300)
plt.show()

# 4. Logical Validation for Thesis
if p_val < 0.05:
    print("\nCONCLUSION: H1 Validated. Statistically significant relationship found.")
else:
    print("\nCONCLUSION: Fail to reject null hypothesis.")
```
